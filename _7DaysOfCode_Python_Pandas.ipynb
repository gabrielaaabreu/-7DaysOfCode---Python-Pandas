```python
listaTemas = []
for item in emprestimos['localizacao']:
  if item >= 0 and item < 100:
    listaTemas.append('Generalidades. Ciência e conhecimento.')
  elif item >= 100 and item < 200:
    listaTemas.append('Filosofia e psicologia.')
  elif item >= 200 and item < 300:
    listaTemas.append('Religião.')
  elif item >= 300 and item < 400:
    listaTemas.append('Ciências sociais.')
  elif item >= 400 and item < 500:
    listaTemas.append('Classe vaga. Provisoriamente não ocupada.')
  elif item >= 500 and item < 600:
    listaTemas.append('Matemática e ciências naturais.')
  elif item >= 600 and item < 700:
    listaTemas.append('Ciências aplicadas.')
  elif item >= 700 and item < 800:
    listaTemas.append('Belas artes.')
  elif item >= 800 and item < 900:
    listaTemas.append('Linguagem. Língua. Linguística.')
  elif item >= 900 and item < 1000:
    listaTemas.append('Geografia. Biografia. História.')
  else:
    listaTemas.append('Sem tema.')

emprestimos.insert(12, column = 'tema', value = listaTemas)
```

```python
emprestimos['data_emprestimo'] = pd.to_datetime(emprestimos['data_emprestimo'])
emprestimos['data_devolucao'] = pd.to_datetime(emprestimos['data_devolucao'])
emprestimos['data_renovacao'] = pd.to_datetime(emprestimos['data_renovacao'])
emprestimos.head()
```

```python
grafico1 = emprestimos.groupby(emprestimos['data_emprestimo'].dt.year)[['id_exemplar', 'id_emprestimo']].count().plot(y='id_exemplar', kind= 'bar', title='Número de livros emprestados por ano', legend=False)
grafico1.set_xlabel('')
grafico1.set_ylabel('')
```

O ano com maior número de livros emprestados foi 2013, seguido de uma pequena queda em 2014 e estabilidade nos anos seguinte. A partir de 2018, a tendência de queda é mais acentuada.

```python
grafico2 = emprestimos.groupby(emprestimos['data_emprestimo'].dt.month)[['id_exemplar']].count().plot(y = 'id_exemplar', kind='line', title='Número de livros emprestados por mês entre 2010 e 2020', legend=False)
grafico2.set_xlabel('Mês')
grafico2.set_ylabel('')
```

Os meses de maior demanda são março e agosto, enquanto os de menor demanda são janeiro, julho e dezembro. Deste modo, as férias dos funcionários podem ser concentradas nesses meses.

```python
grafico3 = emprestimos.groupby(emprestimos['data_emprestimo'].dt.hour)[['id_exemplar']].count().plot(y = 'id_exemplar', kind='bar', title='Número de livros emprestados por horário', legend=False)
grafico3.set_xlabel('Hora do dia')
grafico3.set_ylabel('')
```

Os horários de maior demanda de alunos são no final da manhã, entre 10 e 11 horas, e no final da tarde, entre 16 e 19 horas. As atividades programadas pela diretoria poderiam ser alocadas no início da manhã, no início da tarde ou após as 19h, quando há menor demanda de atendimento a usuários.

```python
def tabelaFreq(coluna):
  a = emprestimos[coluna].value_counts()
  b = emprestimos[coluna].value_counts(normalize=True)*100
  bArredondado = round(b, 2)
  c = pd.concat([a, bArredondado], axis=1)
  return c
```

```python
tabelaFreq('tipo_vinculo_usuario')
```

```python
tabelaFreq('colecao')
```

```python
tabelaFreq('biblioteca')
```

```python
tabelaFreq('tema')
```

A partir das tabelas geradas pode-se perceber que a maioria dos empréstimos são realizados por alunos de graduação (78,35%), a partir do acervo circulante (99,07%), na Biblioteca Central Zila Mamede (68,81%) e são de livros de ciências aplicadas (63,05%). Assim, a diretoria pode investir na ampliação do acervo da Biblioteca Central Zila Mamede e em ações para promover o uso dos outros acervos disponíveis.
