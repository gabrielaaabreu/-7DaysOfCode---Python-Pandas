```python
listaTemas = []
for item in emprestimos['localizacao']:
  if item < 100:
    listaTemas.append('Generalidades. Ciência e conhecimento.')
  elif item >= 100 and item < 200:
    listaTemas.append('Filosofia e psicologia.')
  elif item >= 200 and item < 300:
    listaTemas.append('Religião.')
  elif item >= 300 and item < 400:
    listaTemas.append('Ciências sociais.')
  elif item >= 400 and item < 500:
    listaTemas.append('Classe vaga. Provisoriamente não ocupada.')
  elif item >= 500 and item < 600:
    listaTemas.append('Matemática e ciências naturais.')
  elif item >= 600 and item < 700:
    listaTemas.append('Ciências aplicadas.')
  elif item >= 700 and item < 800:
    listaTemas.append('Belas artes.')
  elif item >= 800 and item < 900:
    listaTemas.append('Linguagem. Língua. Linguística.')
  elif item >= 900 and item < 1000:
    listaTemas.append('Geografia. Biografia. História.')
  else:
    listaTemas.append('Sem tema.')

emprestimos.insert(12, column = 'tema', value = listaTemas)
```
